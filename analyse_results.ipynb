```python
import pandas as pd
import json
import os
```

```python
pd.set_option("display.max.columns", None)
pd.set_option("display.precision", 2)
```

```python
frames = []
for root, dirs, files in os.walk("results"):
   for name in files:
      if name.endswith(".json"):
          frame = pd.read_json(os.path.join(root, name), orient="record", lines=True)
          frames.append(frame)
results = pd.concat(frames)
```

```python
columns = ['task_id',  'problem_type', 'f1', 'recall', 'precision', 'roc', 'oversampling_enabled', 'ratio_after_oversampling', 'undersampling_enabled', 'ratio_after_undersampling', 'svm_kernel', 'svm_predict_probability', 'svm_class_weight', 'cm', 'cm_normalized', 'n_estimators', 'learning_rate', 'subsample', 'embedding_vecor_length', 'epochs', 'batch_size', 'num_lstm_cells', 'dropout_emb_lstm', 'dropout_lstm_dense', 'n_trees_in_forest', 'max_features', 'input_src_path', 'max_vocab_size', 'window_size', 'step_size', 'encode_type', 'test_split_percentage', ]
results = results[columns]
results.head()
```

# Return Null

```python
return_none = results[results["problem_type"] == "RETURN_NULL"]
return_none.sort_values(by="f1", ascending=False).head()
```
