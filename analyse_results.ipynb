```python
import pandas as pd
import json
import os
```

```python
pd.set_option("display.max.columns", None)
pd.set_option("display.precision", 4)
```

```python
frames = []
for root, dirs, files in os.walk("results/final_run"):
   for name in files:
      if name.endswith(".json"):
          frame = pd.read_json(os.path.join(root, name), orient="record", lines=True)
          frames.append(frame)
results = pd.concat(frames)
results.columns
```

```python
columns = ['task_id',  'problem_type', 'f1', 'recall', 'precision', 'roc', 'train_f1', 'train_recall', 'train_precision', 'train_roc', 'train_dev_f1', 'train_dev_recall', 'train_dev_precision', 'train_dev_roc', 'oversampling_enabled', 'ratio_after_oversampling', 'undersampling_enabled', 'ratio_after_undersampling', 'cm', 'cm_normalized', 'n_estimators', 'learning_rate', 'subsample', 'embedding_vecor_length', 'epochs', 'batch_size', 'num_lstm_cells', 'dropout_emb_lstm', 'dropout_lstm_dense', 'n_trees_in_forest', 'max_features', 'input_src_path', 'max_vocab_size', 'window_size', 'step_size', 'encode_type', 'test_split_percentage']
#,'svm_kernel', 'svm_predict_probability', 'svm_class_weight'
results = results[columns]
results.head()
```

# Return Null

```python
return_none = results[(results["problem_type"] == "RETURN_NULL")  & (results["max_vocab_size"] == 100000)]
return_none.sort_values(by="f1", ascending=False)
```

```python
return_none = results[(results["problem_type"] == "RETURN_NULL")  & (results["max_vocab_size"] == 100000) & (results["n_estimators"].notnull())]
return_none.sort_values(by="f1", ascending=False)
```

# Validtion

```python
frames = []
for root, dirs, files in os.walk("results/final_validation"):
   for name in files:
      if name.endswith(".json"):
          with open(os.path.join(root, name)) as f:
            d = json.loads(f.read())
            frame = pd.json_normalize(d)
            frames.append(frame)
results = pd.concat(frames)
```
